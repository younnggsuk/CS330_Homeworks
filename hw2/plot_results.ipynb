```python
import numpy as np
from matplotlib import pyplot as plt


# Create the data set
score_cin_dict = {
    "Protonet": [
        (1, (98.1, 0.4)),
        (2, (99.6, 0.2)),
        (4, (99.9, 0.1)),
        (6, (100.0, 0.0)),
        (8, (99.9, 0.1)),
        (10, (100.0, 0.0))
    ],
    "MAML": [
        (1, (97.2, 0.3)),
        (2, (98.7, 0.2)),
        (4, (99.1, 0.1)),
        (6, (99.3, 0.1)),
        (8, (99.2, 0.1)),
        (10, (99.4, 0.1))
    ]
}
# Plot the data with confidence intervals(shaded)
plt.figure(figsize=(10, 5))
plt.title("Protonet vs MAML")
plt.gca().set_facecolor("#EAEAF2")
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)
plt.gca().spines["bottom"].set_visible(False)
plt.gca().spines["left"].set_visible(False)
plt.gca().set_axisbelow(True)

plt.xlabel("Number of shots")
shots = [str(x[0]) for x in score_cin_dict["Protonet"]]
plt.xticks(np.arange(len(shots)), shots)

plt.ylabel("Accuracy")
plt.ylim(96.5, 100.5)
plt.yticks(np.arange(96, 101, 1))

plt.grid(color="white", linewidth=1, linestyle="-")
plt.tick_params(axis="x", bottom=False, top=False)
plt.tick_params(axis="y", left=False, right=False)

# Plot the data
line_style_dict = {
    "Protonet": {"linestyle": "solid", "marker": "o", "linewidth": 1, "markersize": 4},
    "MAML": {"linestyle": "dashed", "marker": "s", "linewidth": 1, "markersize": 4}
}
for model in score_cin_dict.keys():
    scores = np.array([x[1][0] for x in score_cin_dict[model]])
    errors = np.array([x[1][1] for x in score_cin_dict[model]])
    plt.plot(shots, scores, label=model, **line_style_dict[model])
    plt.fill_between(shots, scores - errors, scores + errors, alpha=0.3)
plt.legend(loc="lower right")
plt.show()
```
